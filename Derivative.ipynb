```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 9, 10) # abscissa

def fx(x):
    return x**2 / 10

""" f'(x) = 0.2 * x 
    tangent y - y0 = f'(x0)*(x - x0)
    perpend y - y0 = (-1 / f'(x0))*(x - x0) """


x_zeros = (1, 2, 4, 6, 8)
y_zeros = list(map(fx, x_zeros))

f, ax = plt.subplots(figsize=(10, 10))
ax.set_aspect('equal')
ax.set_ylim(-10, 10)
plt.plot(x, fx(x), linewidth=4)
plt.scatter(x_zeros, y_zeros, marker="d", s=80, c="r")
for x_z, y_z in zip(x_zeros, y_zeros):
    y_plot_1 = []
    y_plot_2 = []
    for x_i in x:
        y = (0.2 * x_z) * (x_i - x_z) + y_z
        y_p = (-1 / (0.2 * x_z)) * (x_i - x_z) + y_z
        y_plot_1.append(y)   
        y_plot_2.append(y_p)
    plt.plot(x, y_plot_1)
    plt.plot(x, y_plot_2, linestyle=':')
```
